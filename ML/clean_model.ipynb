```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
```

```python
model_df = pd.read_csv("../data/model.csv")
model_df
```

```python
model_df = model_df[["FL_DATE", "WEATHER_DELAY", "DurationofEvent"]]
model_df.head(40)
```

```python
event_df = model_df["DurationofEvent"].tolist()
event_df
```

```python
event_df = [0 if x != x else x for x in event_df]
event_df
```

```python
model_df["DurationofEvent"] = event_df
model_df
```

```python
clean_model = model_df.dropna()
clean_model
```

```python
clean_model.rename(columns={"DurationofEvent": "LENGTH_OF_THUNDERSTORM"},inplace=True)
```

```python
clean_model["NEW_WEATHER"] = clean_model["WEATHER_DELAY"].apply(lambda x: 1 if x > 0 else 0)
```

```python
clean_model
```

```python
clean_model["NEW_THUNDERSTORM"] = clean_model["LENGTH_OF_THUNDERSTORM"].apply(lambda x: 0 if x == 0 else 1)
```

```python
clean_model
```

```python
X = clean_model.drop("NEW_WEATHER", axis=1)
y = clean_model["NEW_THUNDERSTORM"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```
