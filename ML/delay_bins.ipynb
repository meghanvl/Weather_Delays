```python
# Importing Dependencies
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
```

```python
# Reading CSV and making DataFrame
f = pd.read_csv("../data/new_model.csv")
model_df = pd.DataFrame(f)
```

```python
# Grabbing the Columns Needed
duration_df = model_df[["FL_DATE", "WEATHER_DELAY", "DurationinMinutes"]]
duration_df.head()
```

```python
# Dropping NaN
duration_df.dropna(inplace=True)
```

```python
# Renaming Column
duration_df.rename(columns={"DurationinMinutes": "LENGTH_OF_THUNDERSTORM"},inplace=True)
```

```python
## Creating Binary Data
# duration_df["NEW_WEATHER"] = duration_df["WEATHER_DELAY"].apply(lambda x: 1 if x > 0 else 0)
# duration_df["NEW_THUNDERSTORM"] = duration_df["LENGTH_OF_THUNDERSTORM"].apply(lambda x: 0 if x == 0 else 1)
```

```python
# Changing Int to Float
duration_df["LENGTH_OF_THUNDERSTORM"] = duration_df["LENGTH_OF_THUNDERSTORM"].astype(float)
```

```python
# Creating Bins and Labels
bins = [0.0,60.0,120.0,180.0,240.0,300.0,360.0,420.0,480.0,540.0,600.0,660.0,720.0,780.0,840.0,900.0,960.0,1020.0,1080.0,1140.0,1200.0,1260.0,1320.0] 
labels = ["0-60", "61-120", "121-180", "181-240", "241-300", "301-360", "361-420", "421-480", "481-540", "541-600", "601-660",
         "661-720", "721-780", "781-840", "841-900", "901-960", "961-1020", "1021-1080", "1081-1140", "1141-1200", "1201-1260", "1261-1320"]

duration_df["THUNDERSTORM_BINS"] = pd.cut(duration_df["LENGTH_OF_THUNDERSTORM"], bins, labels=labels, include_lowest=True)
duration_df
```

```python
# Group by Bins
group_df = duration_df.groupby("THUNDERSTORM_BINS").mean()
group_df
```

```python
# Dropping NaN
group_df.fillna(0)
```
