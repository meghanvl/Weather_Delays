```python
# Import Dependencies
import pandas as pd
import datetime
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "../data/ORD.csv"
f = pd.read_csv(csv_path)
meater_df = pd.DataFrame(f)
metar_df.tail()
```

```python
metar_df = metar_df[["station", "valid", "lon", "lat", "wxcodes"]].dropna()
```

```python
metar_df = metar_df.loc[(metar_df["wxcodes"] == 'VCTS') | (metar_df["wxcodes"] == 'TS') | 
             (metar_df["wxcodes"] == '+TSRA') | (metar_df["wxcodes"] == '-TSRA BR') | 
             (metar_df["wxcodes"] == 'VCTS BR') | (metar_df["wxcodes"] == 'TS BR') | 
             (metar_df["wxcodes"] == 'VCTS -RA BR')| (metar_df["wxcodes"] == 'VCTS -RA') | 
             (metar_df["wxcodes"] == 'VCTS RA') | (metar_df["wxcodes"] == 'VCTS +RA BR') |
             (metar_df["wxcodes"] == '+TSRA FG') | (metar_df["wxcodes"] == 'VCTS RA BR') |
             (metar_df["wxcodes"] == 'TS HZ') | (metar_df["wxcodes"] == 'VCTS +RA FG') |
             (metar_df["wxcodes"] == 'TSRA BR') | (metar_df["wxcodes"] == '+TSRA BR') |
             (metar_df["wxcodes"] == '-TSRA') | (metar_df["wxcodes"] == 'TSRA')]
```

```python
split_data = metar_df["valid"].str.split(" ")
data = split_data.to_list()
names = ["Date", "Time"]
new_df = pd.DataFrame(data, columns=names)
new_df['valid'] = metar_df['valid'].tolist()

merge_df = metar_df.merge(new_df, how='inner', on='valid')
merge_df
```

```python
# merge_df['Time'] = pd.to_timedelta(merge_df['Time']+':00')
# merge_df
merge_df['valid'] = pd.to_datetime(merge_df['valid'])
```

```python
merge_df['Time'] = pd.to_timedelta(merge_df['Time']+':00')
merge_df['Date'] = merge_df['valid'].dt.date
```

```python
merge_df['Time']
```

```python
event_date = []
durationList = []
startTime = pd.to_timedelta('00:00:00')
endTime = pd.to_timedelta('00:00:00')
for i in range(len(merge_df['Time'])-1):
    if i == 0:
        startTime = merge_df['Time'][i]
    if merge_df['Date'][i + 1] != merge_df['Date'][i]:
        endTime = merge_df['Time'][i]
        timeChange = endTime - startTime
        durationList.append(timeChange)
        event_date.append(merge_df['Date'][i])
        startTime = merge_df['Time'][i+1]
```

```python
duration_df = pd.DataFrame({'FL_DATE': event_date, 'Duration of Event': durationList})
```

```python
duration_df['Duration of Event'].unique()
```

```python
total_seconds_list = []
for i in range(len(duration_df['Duration of Event'])):
    total_seconds_list.append(duration_df['Duration of Event'][i].total_seconds())
```

```python
duration_df.to_csv("../data/duration.csv")
```
